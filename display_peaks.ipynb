```python
import pandas as pd
import matplotlib.pyplot as plt

# Path to the CSV file
csv_filename = "energy_peaks.csv"

# Load the CSV into a pandas DataFrame
try:
    df = pd.read_csv(csv_filename, parse_dates=["timestamp"])
except FileNotFoundError:
    print(f"{csv_filename} not found.")
    exit()
except Exception as e:
    print(f"Error reading {csv_filename}: {e}")
    exit()

# If you have a lot of timestamps, consider plotting index instead for readability
x = df['timestamp']
y = df['peak_value']

plt.figure(figsize=(10, 4))
markerline, stemlines, baseline = plt.stem(x, y)
plt.setp(markerline, markersize=4)
plt.xlabel("Time (UTC)")
plt.ylabel("Peak Value (normalized)")
plt.title("Stem Plot of Detected Energy Peaks Over Time")
plt.xticks(rotation=30)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
